# 연산자
- 연산자는 산술, 논리, 비교 등의 계산을 수행하는 Python의 특수 기호

## 산술연산자
- 산술연산의 결과를 반환한다.
- `+, -, *, / , //, %, **`

## 대입 연산자
- 값(객체)을 변수에 할당
- 산술연산결과의 값을 변수에 할당
    - `= , += , -= , ...`

## 비교 연산자
- 변수 또는 값(객체)을 비교하여 bool 객체를 반환
- `> , < , >=, <= , !=, ==`

## 논리연산자
- 논리 자료형인 bool객체를 연산하여 bool 객체를 반환
- and, or, not
    - A and B : A와 B 모두 True 면 결과는 True, 아니면 False
    - A or B : A와 B 둘 중 하나만 True여도 결과는 True , 둘다  False면 False
    - not A : A가 True면 False가 되고, False면 True가 된다.


## 포함 연산자
- 어떠한 값이 포함 되어있는지 여부에 따라 bool 객체를 반환


```python
"a" in "abcd"
```

## is 와 is not 연산자
- 동일한 주소에 할당된 객체임을 비교하는 연산자로 bool 객체를 반환
- `is`: 동일한 주소에 할당된 객체이면 True, 아닐경우 False
- `is not`: 동일한 주소에 할당된 객체이면 False, 아닐경우 True

1. 3명의 학생의 시험점수가 각 변수에 들어가있다. 변수의 값들을 이용하여 평균을 구하고 구한 평균을 이용하여 표준 편차를 구하시오.

$$
\mu = \frac{1}{N}\sum_{i=1}^{N}x_i
$$
$$
\sigma^{2} = \frac{1}{N}\sum_{i=1}^{N}(x_i - \mu)^{2}
$$

```python
karns = 90
minsu = 87
gildong = 72

#평균
average = (karns+minsu+gildong)/3
#분산
dispersion = ((average-karns)**2 + (average-minsu)**2 + (average-gildong)**2)/3
#표준편차
standard_dispersion = dispersion**0.5

print(average)
print(dispersion)
print(standard_dispersion)
```
