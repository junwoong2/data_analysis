# 문자열이란?
- 여러 개의 문자를 순서대로 나열한 자료형(`str`)
- Sequence 자료형
- immutable 객체
- 컴퓨터는 문자도 숫자(이진수)로 바꿔서 처리한다.
- 컴퓨터가 문자를 처리하기 위해 이진수로 변환되는 표준규칙을 인코딩이라 한다.
- ASCII, EUC-KR , UTF-8 등등
- https://namu.wiki/w/%EC%95%84%EC%8A%A4%ED%82%A4%20%EC%BD%94%EB%93%9C

## 이스케이프코드(escape code)
- 키보드로 직접입력이 불가능해 미리 정의해 둔 문자조합
- `\n` : 줄바꿈
- `\t` : 탭키       

```python
text = "hello\nmy name is joonwoong"
print(text)
```

## Doc String
- 따옴표를 3개를 이어서 쓰면 된다.
> `'''문자열'''`, `"""문자열"""`
- Doc String 안에 위치한 문자열은 무조건 적은대로 표현
- 여러줄 주석으로도 사용함

## 문자열 포매팅(String Formatting)
- 문자열 내의 특정한 값을 변수를 이용해 삽입하는 방법
- 포맷 코드
    - `%s`: 문자열(str)
    - `%d`: 정수(int)
    - `%f`: 실수(float)
