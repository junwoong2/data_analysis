# 반복 가능한 객체(iterable)
- 말 그대로 반복할 수 있는 객체를 말함
- 즉, 요소가 여러 개 있어서, 차례대로 반환할 수 있는 객체
- `list, dict, set, str, tuple, range` . . .

# 이터레이터(iterator)
- 반복 가능한 객체를 생성하는 방법으로 값을 차례대로  반환할 수 있는 객체
- 이터레이터는 `for` 문이나 `next` 함수를 통해 값을 하나씩 얻을 때 사용

````python
lst = [1,2,3]
iterator = iter(lst)
iterator
````

````python
for n in lst:
    print(n)
````

````python
for n in iterator:
    print(n)
````

````python
zip_obj = zip(lst, lst)
````

````python
for n1, n2 in zip_obj:
    print(n1,n2)
````

- next 함수
    - iterator 객체를 동작시켜 차례대로 반환할 수 있는 함수
    - 더 이상 꺼낼 수 없으면 에러 발생!!

````python
iterator = iter(lst)
````

````python
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))#에러발생
````

# 제너레이터(generator)
- 이터레이터의 한 종류
- iterator 와 비슷한 동작을 하는 객체
- `yield` 키워드를 사용하는 함수를 실행하면, 그 함수는 Generator 객체를 반환
- `yield`
    - 값을 반환하고 현재 상태를 저장

- 예시

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

````python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
type(gen)
````

````python
next(gen)
````

````python
def gen_func(n):
    for i in range(n):
        yield i
````

````python
gen = gen_func(10)
````

````python
for i in gen:
    print(i)
````

````python
gen = gen_func(10000000)
lst = list(range(10000000))
````

````python
import sys
sys.getsizeof(gen), sys.getsizeof(lst)
````

````python
%%timeit
r = []
for data in lst:
    r.append(data)
````

````python
%%timeit
r = []
for data in gen:
    r.append(data)
````

````python
gen = (x for x in range(3,300,3))

for i in gen:
    print(i)
````

````python
def gen_func(lst):
    yield from lst
````

````python
gen = gen_func([1,2,3])
````

````python
def gen_enumerate(lst):
    length = len(lst)
    for i in range(length):
        yield i, lst[i]
````

````python
gen = gen_enumerate([1,2,3])
gen
````
