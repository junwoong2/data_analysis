## 1. MAE 와 RMSE 함수 구현하기
- 다음과 같이 실제 값과 , 모델이 예측한 값이 각각 100개가 있다.
- 실제값과 예측값 사이에 평균 오차를 구하는 함수 두개를 구현해보자.

````python
np.random.seed(42)
target = np.random.randint(50, 91,100) # 정답값
pred = np.random.randint(40, 100,100) # 예측값
target , pred
````

### MAE(Mean Absolute Error)
- 실제 값과 예측 값의 차이를 절대값으로 변환해 평균
$$
MAE = \frac{1}{n}{\sum_{i=1}^{n}|y_i-\hat{y}_i|}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
def mae(target,pred):
    return abs(np.subtract(target, pred))
    
print(mae(target,pred))
````

### RMSE(Root Mean Squared Error)
- 실제 값과 예측 값의 차이를 제곱한 뒤 평균한 후에 루트

$$
RMSE = \sqrt{\frac{1}{n}{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
def rmse(target,pred):
    sum = np.square(np.subtract(target, pred))
    return np.sqrt((np.sum(sum)/len(sum)))

rmse(target, pred)
````

````python
def split_dataset(dataset):
    total_idx = np.arange(len(dataset))
    slice_20 = int(len(dataset) * 0.2)
    idx_list = []

    for i in range(5):
        # 차집합 이용해서 뽑힌 인덱스 제외하고 unchosen에 저장
        # unchosen = np.setdiff1d(total_idx, np.concatenate(idx_list)) : 비어있는 리스트 안합쳐져서 조건 추가
        if idx_list:
            unchosen = np.setdiff1d(total_idx, np.concatenate(idx_list))
        else:
            unchosen = total_idx

        # 마지막 인덱스일 경우 그냥 전부를 랜덤 돌리기
        if i == 4:
            idx_list.append(np.random.choice(unchosen, len(unchosen), replace = False))
        else:
            idx_list.append(np.random.choice(unchosen, slice_20, replace=False))

    return idx_list

print(split_dataset(x_train))
````

- 다음 셀은 위에 함수 실행 결과를 검증하는 부분

````python
index_list = split_dataset(x_train)
lst = []
for idx in index_list:
    lst.extend(idx)
np.all( np.array(sorted(lst)) == np.arange(len(x_train)) ) # True 가 나와야 성공!!
````
